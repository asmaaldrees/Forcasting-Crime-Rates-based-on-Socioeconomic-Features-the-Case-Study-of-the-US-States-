```python
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
# Set notebook mode to work in offline
pyo.init_notebook_mode(connected=True)
import plotly.io as pio
pio.renderers.default = 'iframe'
import plotly.express as px
```

```python
Crime=pd.read_csv("csvData.csv")
```

```python
Crime
```

```python
geo=pd.read_csv("GEO.csv")
```

```python
geo['Place Name']=geo['Place Name'].str.split(",",expand=True)[0]
```

```python
merged_inner = pd.merge(left=Crime, right=geo, left_on='State', right_on='Place Name')
```

```python
merged_inner
```

```python


px.scatter_mapbox(merged_inner, lat="Latitude", lon="Longitude", hover_name="State",
                  color="rate",size="rate",title="Crime Rate in US States").update_layout(
    mapbox={
        "style": "carto-positron",
        "zoom": 3.5
    }
)
```

```python


px.scatter_mapbox(merged_inner, lat="Latitude", lon="Longitude", hover_name="State",
                  color="rate",size="violentRate",title="Violent Crime Rate in US States").update_layout(
    mapbox={
        "style": "carto-positron",
        "zoom": 3.5
    }
)
```

```python
Soc_Data=pd.read_csv("Final_Selected_Features.csv")
```

```python
merged_inner = pd.merge(left=Soc_Data, right=merged_inner, left_on='States', right_on='Place Name')
```

```python
merged_inner=merged_inner.dropna(axis=1)
```

```python
merged_inner
```

```python
I=4
import time
while I<len(merged_inner.columns)-11:
    try:
        print(merged_inner.columns[I])


        fig=px.scatter_mapbox(merged_inner, lat="Latitude", lon="Longitude", hover_name="State",
                          color=merged_inner.columns[I],size=merged_inner.columns[I],title=str(merged_inner.columns[I])+" Rate in US States").update_layout(
            mapbox={
                "style": "carto-positron",
                "zoom": 3.5
            }
        )
        fig.write_html(str(merged_inner.columns[I])+".html")
        fig.show()
        time.sleep(5)
    except:
        pass
    I=I+1
```

```python
merged_inner=merged_inner.sort_values("rate").reset_index(drop=True)
fig = px.bar(
    merged_inner, x="Place Name", y='rate',title="Crime Rates in US States",color="rate")
fig.show()
```

```python
import matplotlib.pyplot as plt
y = merged_inner['Percent!!RACE!!Total population!!One race!!Asian!!Filipino']
z = merged_inner['rate']
n = merged_inner['Place Name']

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(z, y,)


for i, txt in enumerate(n):
    ax.annotate(txt, (z[i], y[i]))
plt.title("Crime vs Asian and Filipino Populations")
plt.xlabel("Crime Rates")
plt.ylabel("Asian and Filipino Populations")
plt.show()
```
